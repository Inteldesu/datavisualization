2. Dataset Exploration

In this section, the dataset is loaded using Pandas and basic information is displayed to understand its structure.

```python
import pandas as pd

df = pd.read_csv("student_alcohol_consumption_300x5.csv")
df.head()
```

```python
df.info()
```

 Descriptive Statistics

Selected Variable: Age  

In this section, we compute the mean, median, mode, standard deviation, minimum, and maximum values of the selected numerical variable.

```python
print("Mean:", df["age"].mean())
print("Median:", df["age"].median())
print("Mode:", df["age"].mode()[0])
print("Standard Deviation:", df["age"].std())
print("Minimum:", df["age"].min())
print("Maximum:", df["age"].max())
```


1. The average value of the variable (age) is shown by the mean, which represents the typical age of students in the dataset.  

2. If the mean is close to the median, it indicates that the data is fairly symmetric and not heavily skewed.  

3. The standard deviation shows how spread out the values are. A higher value means greater variability, while a lower value indicates that the data points are closer to the mean.  

 4. Frequency Distribution

Selected Variable: Walc (Weekend Alcohol Consumption)

This section shows the frequency and percentage distribution of the selected categorical variable.

```python
freq = df["Walc"].value_counts()
perc = df["Walc"].value_counts(normalize=True) * 100

pd.DataFrame({
    "Frequency": freq,
    "Percentage (%)": perc
})
```


- The most frequent category is the level of alcohol consumption with the highest count in the table.  
- The least frequent category is the level with the lowest count.  
- The distribution suggests how common or rare certain alcohol consumption behaviors are among students. If one category dominates, it indicates that most students share similar drinking habits.

5. Univariate Data Visualizations

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["age"], bins=10)
plt.title("Histogram of Age")
plt.show()
```



- The shape of the distribution shows whether the data is symmetric or skewed.  
- Most observations are concentrated around the age range with the highest bars in the histogram.  

```python
sns.boxplot(x=df["age"])
plt.title("Box Plot of Age")
plt.show()
```


- Outliers are shown as points outside the whiskers of the box plot.  
- The IQR (interquartile range) represents the middle 50% of the data, showing how spread out the central values are.  

```python
sns.kdeplot(df["studytime"], fill=True)
plt.title("KDE Plot of Study Time")
plt.xlabel("Study Time")
plt.ylabel("Density")
plt.show()
```


- The highest density indicates where most values are concentrated.  
- The distribution may appear symmetric or skewed depending on the curve shape.  
- A single peak indicates a unimodal distribution, while multiple peaks indicate multimodality.  

```python
sns.countplot(x="Walc", data=df)
plt.title("Bar Chart of Walc")
plt.show()
```


- The category with the tallest bar has the highest frequency.  
- The category with the shortest bar has the lowest frequency.  

```python
df["Walc"].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Pie Chart of Walc")
plt.ylabel("")
plt.show()
```


- The pie chart shows the proportion of each category.  
- If one section is much larger, the dataset is dominated by that category. Otherwise, it is more balanced.
